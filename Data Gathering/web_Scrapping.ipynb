```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}
webPage = requests.get('https://www.ambitionbox.com/list-of-companies?page=1', headers=headers).text
```

```python
name =[]
rating = []
founded = []
india_employee = []
global_employee = []
headquarters = []
website = []
primary_industry = []
```

```python
web_soup = BeautifulSoup(webPage, 'lxml')
```

```python
company = web_soup.find_all('h2', class_ = 'companyCardWrapper__companyName')
for i in company:
    name.append(i.get_text(strip=True))

len(name)
```

```python
for i in web_soup.find_all('div', class_ = 'rating_text'):
    rating.append(i.text.strip())
    
len(rating)
```

```python
base_url = "https://www.ambitionbox.com/overview/{}-overview" 
```

```python
for i in name:
    url = base_url.format(i.lower().replace(" ", "-"))
    overview = requests.get(url, headers=headers).text
    soup = BeautifulSoup(overview, "lxml")

        # Company Age
    age_spans = soup.find_all('span', class_='css-1jxf684 text-primary-text text-sm font-pn-600 flex-[6] md:flex-[auto]')
    founded.append(age_spans[0].text if age_spans else 'N/A')
```

```python
print(name )
print(rating )
print(founded )
print(global_employee )
```
