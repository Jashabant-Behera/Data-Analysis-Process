### Types of Data Assessment

There are two types of Assessment:

1. `Manual` : Looking through the data manually to identify the issues in the dataset.

2. `Programmatic` : By using pandas to identify the issues in the dataset.
    - head and tail
    - sample
    - isnull
    - deplicated
    - describe
    - info
    


There are two steps involved in Assessment:

1. **Export Data** : Export the data to an Excel or Pandas Dataframe.



2. **Note Down the Report**: Document the issues found during the manual acessment. Categprise the issue into two types:

    - `Dirty Data` : Inconsistent, incorrect or incomplete data.
    - `Messy Data` : Data that are not organized properly.

### `Note` : Assessing the Data is a Continous and Iterative Process

### Data Quality Dimentions
1. `Completeness`: Ensure all require data is present. Is Data Missing?

2. `Validity`: Ensure data conforms to the defined formats and rules. Is Data Valid?

3. `Accuracy`: Ensure data is correct and free from errors. Is Data Accurate?

4. `Consistency`: Ensure data is consistent across different datasets and sources. Is Data Consistent?


### `Note` : Categorise the data written in the report based on these dimentions.

### Order of Severity
Completeness <- Validity <- Accuracy <- Consistency

### Data Cleaning:

* Data Cleaning Order:

    1. Quality -> Completeness
    2. Tidiness
    3. Quality -> Validity
    4. Quality -> Accuracy
    5. Quality -> Consistency

* Steps involved in Data Cleaning:

    1. `Define`: Define the Solution for each issue identified in the assessment report.
    2. `Code`: Implement the solution using Pandas or any other data manipulation library.
    3. `Test`: Verify that the solution has resolved the issue without introducing new problems.

### `Note` : Before starting the data cleaning, make a copy of the original dataset to avoid any loss of data.
